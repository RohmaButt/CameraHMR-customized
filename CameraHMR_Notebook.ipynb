```python
import cv2
import os

def extract_frames(video_path, output_folder, start_sec=0, end_sec=None):
    """
    Extracts frames from a video between start_sec and end_sec and saves them as images.

    Args:
        video_path (str): Path to the input video file.
        output_folder (str): Path to the folder where extracted frames will be saved.
        start_sec (int or float): Start time in seconds to begin extracting frames.
        end_sec (int or float): End time in seconds to stop extracting frames.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    vidcap = cv2.VideoCapture(video_path)
    fps = vidcap.get(cv2.CAP_PROP_FPS)
    total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))

    start_frame = int(start_sec * fps)
    end_frame = int(end_sec * fps) if end_sec is not None else total_frames

    # Set video to start frame
    vidcap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

    count = start_frame
    saved_count = 0
    while count < end_frame:
        success, image = vidcap.read()
        if not success:
            break

        frame_filename = os.path.join(output_folder, f"frame_{count:06d}.jpg")
        cv2.imwrite(frame_filename, image)
        saved_count += 1
        if saved_count % 100 == 0:
            print(f"Processed {saved_count} frames.")

        count += 1

    vidcap.release()
    print(f"Finished extracting {saved_count} frames to {output_folder} from {start_sec}s to {end_sec}s.")

# Example usage:
video_input_path = '/content/CameraHMR-customized/demo_images/run_close.mp4'
frames_output_folder = '/content/CameraHMR-customized/demo_frames_run'
extract_frames(video_input_path, frames_output_folder, start_sec=0, end_sec=5)
```

```python
import os
import re

def find_and_sort_overlay_images(folder_path):
    """
    Finds and sorts images with the suffix '_clean_overlay' in a given folder.

    Args:
        folder_path (str): The path to the folder to search.
    """
    overlay_images = []
    pattern = re.compile(r'^(.*)_clean_overlay\.(jpg|png)$')

    if not os.path.exists(folder_path):
        print(f"Error: Folder not found at {folder_path}")
        return

    for filename in os.listdir(folder_path):
        match = pattern.match(filename)
        if match:
            overlay_images.append(filename)

    # Sort the images alphabetically (which will sort by frame number if named consistently)
    overlay_images.sort()

    if overlay_images:
        print(f"Found {len(overlay_images)} images with '_clean_overlay' suffix in {folder_path}:")
        for img_name in overlay_images:
            print(img_name)
    else:
        print(f"No images with '_clean_overlay' suffix found in {folder_path}.")

# Example usage: Replace with your actual folder path
# You can use the output folder from the previous steps, e.g.,
# '/content/drive/My Drive/Sheheryar/output_video_results_run_close'
# or '/content/CameraHMR-customized/output_frames' if you kept files there

folder_to_search = '/content/drive/My Drive/Sheheryar/output_video_results_run_close' # Change this to your output folder
find_and_sort_overlay_images(folder_to_search)
```

```python
import cv2
import os
import re

# Folder containing the sorted images from the previous step
image_folder = '/content/drive/My Drive/Sheheryar/output_video_results_run_close'  # Make sure this matches your output folder

# Output video file path
output_video_path = './output_video_overlay.mp4'

# Frame rate for the output video
fps = 25

# Regex to extract frame number from filename for sorting
# This pattern assumes filenames are like 'frame_XXXXXX_YYYYYY_clean_overlay.jpg'
pattern = re.compile(r'frame_(\d+)_.*_clean_overlay\.(jpg|png)')

# Collect and sort images by the first frame number (XXXXXX)
image_files = []
if os.path.exists(image_folder):
    for filename in os.listdir(image_folder):
        match = pattern.match(filename)
        if match:
            frame_num = int(match.group(1)) # Extract the first number group
            image_files.append((frame_num, filename))

# Sort by frame number
image_files.sort(key=lambda x: x[0])

# Print sorted list of image files to be used in the video
print("📋 Sorted images to be used in video:")
if image_files:
    for frame_num, filename in image_files:
        print(f"  Frame {frame_num:06d}: {filename}")
else:
    print(f"⚠️ No images with '_clean_overlay' suffix found in {image_folder}. Exiting.")
    exit()


# Load the first image to get dimensions
first_image_path = os.path.join(image_folder, image_files[0][1])
frame = cv2.imread(first_image_path)

if frame is None:
    print(f"❌ Could not read the first image: {first_image_path}. Exiting.")
    exit()

height, width, _ = frame.shape

# Set up video writer
# Adjust codec based on compatibility (e.g., 'mp4v' for mp4, 'XVID' for avi)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
video_writer = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# Write each image into the video
print(f"\nCreating video '{output_video_path}' with {len(image_files)} frames at {fps} fps...")
for i, (_, filename) in enumerate(image_files):
    img_path = os.path.join(image_folder, filename)
    img = cv2.imread(img_path)
    if img is not None:
        video_writer.write(img)
        if (i + 1) % 50 == 0:
            print(f"  Processed {i + 1}/{len(image_files)} frames...")
    else:
        print(f"⚠️ Skipped unreadable file: {filename}")

# Release the video writer
video_writer.release()

print(f"\n✅ Video saved successfully at: {output_video_path}")
```

```python
import os
import re
import shutil
from collections import defaultdict

# Paths
source_folder = "/content/drive/My Drive/output_video_make6"  # Replace with actual folder path
destination_folder = "./selected_per_frame"
os.makedirs(destination_folder, exist_ok=True)

# Regex to parse filename
pattern = re.compile(r"(frame_(\d+))_\d+_person_(\d+)_combined_all_persons\.jpg")

# Dictionary: {frame_number: (person_id, filename)}
frame_to_best_image = {}

for filename in os.listdir(source_folder):
    match = pattern.match(filename)
    if match:
        frame_str = match.group(1)       # e.g., 'frame_000005'
        frame_num = int(match.group(2))  # e.g., 5
        person_id = int(match.group(3))  # e.g., 8

        # Check if this frame already has a selected person image
        if frame_str not in frame_to_best_image:
            frame_to_best_image[frame_str] = (person_id, filename)
        else:
            existing_person_id, _ = frame_to_best_image[frame_str]
            if person_id > existing_person_id:  # Keep highest person ID
                frame_to_best_image[frame_str] = (person_id, filename)

# Copy selected images
for _, (_, filename) in frame_to_best_image.items():
    src_path = os.path.join(source_folder, filename)
    dst_path = os.path.join(destination_folder, filename)
    shutil.copy2(src_path, dst_path)
    print(f"Copied: {filename}")

print("✅ Done. One image per frame copied.")
```

```python
# Import dependencies
from google.colab.output import eval_js
import cv2
import numpy as np
import os
import subprocess
import time
from IPython.display import display, Javascript, Image, clear_output
from base64 import b64decode, b64encode
import PIL
import io
from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler
import threading
import queue
from pathlib import Path
import glob
import signal

# Function to convert JavaScript object to OpenCV image
def js_to_image(js_reply):
    image_bytes = b64decode(js_reply.split(',')[1])
    jpg_as_np = np.frombuffer(image_bytes, dtype=np.uint8)
    img = cv2.imdecode(jpg_as_np, flags=1)
    return img

# Function to convert OpenCV image to base64 byte string
def image_to_bytes(img_array):
    _, buffer = cv2.imencode('.png', img_array)
    img_bytes = b64encode(buffer).decode('utf-8')
    return f'data:image/png;base64,{img_bytes}'

# JavaScript for webcam video stream with side-by-side display
def video_stream():
    js = Javascript('''
        var video;
        var div = null;
        var stream;
        var captureCanvas;
        var inputImgElement;
        var outputImgElement;
        var statusElement;

        var pendingResolve = null;
        var shutdown = false;

        function removeDom() {
           if (stream) {
               stream.getVideoTracks()[0].stop();
           }
           if (video) video.remove();
           if (div) div.remove();
           video = null;
           div = null;
           stream = null;
           inputImgElement = null;
           outputImgElement = null;
           statusElement = null;
           captureCanvas = null;
        }

        function onAnimationFrame() {
          if (!shutdown) {
            window.requestAnimationFrame(onAnimationFrame);
          }
          if (pendingResolve) {
            var result = "";
            if (!shutdown) {
              captureCanvas.getContext('2d').drawImage(video, 0, 0, 640, 480);
              result = captureCanvas.toDataURL('image/jpeg', 0.8)
            }
            var lp = pendingResolve;
            pendingResolve = null;
            lp(result);
          }
        }

        async function createDom() {
          if (div !== null) {
            return stream;
          }

          div = document.createElement('div');
          div.style.border = '2px solid black';
          div.style.padding = '10px';
          div.style.width = '100%';
          div.style.maxWidth = '1400px';
          document.body.appendChild(div);

          // Status display
          const statusDiv = document.createElement('div');
          statusDiv.innerHTML = "<span style='font-size: 16px; font-weight: bold;'>Status: </span>";
          statusElement = document.createElement('span');
          statusElement.innerText = 'Initializing...';
          statusElement.style.fontWeight = 'bold';
          statusElement.style.color = 'blue';
          statusDiv.appendChild(statusElement);
          div.appendChild(statusDiv);

          // Create container for side-by-side images
          const imageContainer = document.createElement('div');
          imageContainer.style.display = 'flex';
          imageContainer.style.alignItems = 'top';
          imageContainer.style.gap = '20px';
          imageContainer.style.marginTop = '10px';
          div.appendChild(imageContainer);

          // Left side: Live webcam
          const leftDiv = document.createElement('div');
          leftDiv.style.textAlign = 'center';
          const leftLabel = document.createElement('div');
          leftLabel.innerHTML = '<strong>Live Webcam</strong>';
          leftLabel.style.marginBottom = '5px';
          leftDiv.appendChild(leftLabel);

          video = document.createElement('video');
          video.style.display = 'block';
          video.width = 320;
          video.height = 240;
          video.setAttribute('playsinline', '');
          video.style.border = '1px solid #ccc';
          leftDiv.appendChild(video);
          imageContainer.appendChild(leftDiv);

          // Right side: Processed output
          const rightDiv = document.createElement('div');
          rightDiv.style.textAlign = 'center';
          const rightLabel = document.createElement('div');
          rightLabel.innerHTML = '<strong>Processed Output</strong>';
          rightLabel.style.marginBottom = '5px';
          rightDiv.appendChild(rightLabel);

          outputImgElement = document.createElement('img');
          outputImgElement.style.display = 'block';
          outputImgElement.style.width = '320px';
          outputImgElement.style.height = '240px';
          outputImgElement.style.border = '1px solid #ccc';
          outputImgElement.style.backgroundColor = '#f0f0f0';
          rightDiv.appendChild(outputImgElement);
          imageContainer.appendChild(rightDiv);

          // Instructions
          const instruction = document.createElement('div');
          instruction.innerHTML =
              '<div style="margin-top: 15px; padding: 10px; background-color: #fff3cd; border: 1px solid #ffeaa7; border-radius: 5px;">' +
              '<span style="color: #856404; font-weight: bold;">Batch Processing Instructions:</span><br>' +
              '<span style="color: #856404;">• Processing in batches of 10-15 frames for stability</span><br>' +
              '<span style="color: #856404;">• Script restarts automatically after each batch</span><br>' +
              '<span style="color: #856404;">• First frame may take 30-60 seconds (model loading)</span><br>' +
              '<span style="color: red; font-weight: bold;">• Click anywhere in this area or press Ctrl+C to stop</span>' +
              '</div>';
          div.appendChild(instruction);

          // Click handlers for stopping
          video.onclick = () => { shutdown = true; };
          outputImgElement.onclick = () => { shutdown = true; };
          instruction.onclick = () => { shutdown = true; };

          // Initialize webcam
          try {
            stream = await navigator.mediaDevices.getUserMedia(
                {video: { facingMode: "environment", width: 640, height: 480}});
            video.srcObject = stream;
            await video.play();
          } catch (err) {
            console.error('Error accessing webcam:', err);
            statusElement.innerText = 'Error accessing webcam';
            statusElement.style.color = 'red';
          }

          captureCanvas = document.createElement('canvas');
          captureCanvas.width = 640;
          captureCanvas.height = 480;
          window.requestAnimationFrame(onAnimationFrame);

          return stream;
        }

        async function stream_frame(status, inputImg, outputImg) {
          if (shutdown) {
            removeDom();
            shutdown = false;
            return '';
          }

          stream = await createDom();

          if (status != "") {
            statusElement.innerHTML = status;
          }

          if (outputImg != "") {
            outputImgElement.src = outputImg;
          }

          var result = await new Promise(function(resolve, reject) {
            pendingResolve = resolve;
          });

          return result;
        }
    ''')
    display(js)

def video_frame(status, input_img, output_img):
    data = eval_js('stream_frame("{}", "{}", "{}")'.format(status, input_img, output_img))
    return data

class BatchProcessor:
    def __init__(self, input_folder, logo_path, output_folder, batch_size=12):
        self.input_folder = input_folder
        self.logo_path = logo_path
        self.output_folder = output_folder
        self.batch_size = batch_size
        self.current_process = None
        self.batch_count = 0
        self.total_frames_processed = 0
        self.status_queue = queue.Queue()  # Add this line


    def cleanup_folders(self):
        """Clean up input and output folders"""
        for folder in [self.input_folder, self.output_folder]:
            if os.path.exists(folder):
                for f in os.listdir(folder):
                    file_path = os.path.join(folder, f)
                    try:
                        if os.path.isfile(file_path):
                            os.remove(file_path)
                    except Exception as e:
                        print(f"Warning: Could not remove {file_path}: {e}")

    def kill_existing_processes(self):
        """Kill any existing processing scripts"""
        try:
            # Kill any existing texture_transfer processes
            subprocess.run(['pkill', '-f', 'texture_transfer_metric_depth_transparent_optimized.py'],
                         capture_output=True, text=True)
            time.sleep(2)
        except:
            pass

    def start_processing_script(self):
        """Start a new processing script instance"""
        self.kill_existing_processes()

        command = (
            f'xvfb-run -s "-screen 0 640x480x24" conda run -n camerahmr python3 '
            f'texture_transfer_metric_depth_transparent_optimized.py '
            f'--image_folder {self.input_folder} '
            f'--logo_path {self.logo_path} '
            f'--output_folder {self.output_folder} '
            f'--crop_bottom_percentage 0.0 '
            f'--crop_top_percentage 0.0 '
            f'--logo_size 0.8'
        )

        print(f"\n=== Starting Batch {self.batch_count + 1} ===")
        print(f"Command: {command}")

        # Start the process
        self.current_process = subprocess.Popen(
            command, shell=True,
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            universal_newlines=True,
            bufsize=1,
            preexec_fn=os.setsid  # Create new process group for easier cleanup
        )

        # Start monitoring thread
        monitor_thread = threading.Thread(target=self.monitor_process_output)
        monitor_thread.daemon = True
        monitor_thread.start()

        return self.current_process

    def monitor_process_output(self):
        """Monitor the background process output in a separate thread with frame-by-frame reporting"""
        if not self.current_process:
            return

        current_frame = None
        processing_stage = ""

        while True:
            output = self.current_process.stdout.readline()
            if output == '' and self.current_process.poll() is not None:
                break
            if output:
                line = output.strip()

                # Parse different types of output for frame tracking
                if "Processing frame" in line or "frame_" in line:
                    # Extract frame information
                    import re
                    frame_match = re.search(r'frame_(\d+)', line)
                    if frame_match:
                        current_frame = frame_match.group(1)
                        print(f"🎬 [BATCH-{self.batch_count + 1}] Frame {current_frame}: Starting processing...")

                elif "Loading" in line or "Initializing" in line:
                    processing_stage = "Loading models"
                    print(f"⚙️  [BATCH-{self.batch_count + 1}] {line}")

                elif "depth" in line.lower() or "mesh" in line.lower():
                    processing_stage = "Generating 3D data"
                    if current_frame:
                        print(f"🔍 [BATCH-{self.batch_count + 1}] Frame {current_frame}: {processing_stage}")
                    else:
                        print(f"🔍 [BATCH-{self.batch_count + 1}] {line}")

                elif "texture" in line.lower() or "overlay" in line.lower():
                    processing_stage = "Applying texture/overlay"
                    if current_frame:
                        print(f"🎨 [BATCH-{self.batch_count + 1}] Frame {current_frame}: {processing_stage}")
                    else:
                        print(f"🎨 [BATCH-{self.batch_count + 1}] {line}")

                elif "Saved" in line or "saved" in line or "output" in line.lower():
                    if current_frame:
                        print(f"✅ [BATCH-{self.batch_count + 1}] Frame {current_frame}: Processing complete!")
                        # Update webcam display with completion status
                        try:
                            status = f"Batch {self.batch_count + 1} - Frame {current_frame} completed!"
                            # This would need to be called from main thread, so we'll use a queue
                            self.status_queue.put(status)
                        except:
                            pass
                    print(f"💾 [BATCH-{self.batch_count + 1}] {line}")

                elif "Error" in line or "error" in line or "ERROR" in line:
                    print(f"❌ [BATCH-{self.batch_count + 1}] ERROR: {line}")

                else:
                    # General output
                    print(f"📝 [BATCH-{self.batch_count + 1}] {line}")

            time.sleep(0.01)

    def terminate_current_process(self):
        """Terminate the current processing script"""
        if self.current_process and self.current_process.poll() is None:
            try:
                # Kill the entire process group
                os.killpg(os.getpgid(self.current_process.pid), signal.SIGTERM)
                time.sleep(2)

                # Force kill if still running
                if self.current_process.poll() is None:
                    os.killpg(os.getpgid(self.current_process.pid), signal.SIGKILL)

            except Exception as e:
                print(f"Error terminating process: {e}")

        self.current_process = None

    def final_file_check(self, expected_files):
        """Final check for files after process completion"""
        final_check = []
        for expected_file in expected_files:
            patterns = [
                expected_file,
                os.path.join(self.output_folder, "*.jpg"),
                os.path.join(self.output_folder, "*.png")
            ]
            for pattern in patterns:
                matches = glob.glob(pattern)
                if matches:
                    for match in sorted(matches, key=os.path.getmtime, reverse=True):
                        if ('clean_overlay' in match):
                            if os.path.getsize(match) > 0:
                                final_check.append(match)
                                print(f"📁 Found final output: {os.path.basename(match)}")
                                break
                    break
        return final_check

    # Enhanced wait_for_output method that shows individual frame completion
    def wait_for_output(self, expected_files, timeout=120):
        """Wait for expected output files with timeout and individual frame reporting"""
        timeout_start = time.time()
        completed_frames = set()

        print(f"🔄 Waiting for {len(expected_files)} frames to complete...")

        while time.time() - timeout_start < timeout:
            # Check status queue for updates
            try:
                while not self.status_queue.empty():
                    status = self.status_queue.get_nowait()
                    print(f"📊 Status Update: {status}")
            except:
                pass

            # Check for expected files
            found_files = []
            newly_completed = []

            for i, expected_file in enumerate(expected_files):
                frame_id = f"frame_{i:04d}"

                # Check if this frame was already reported as complete
                if frame_id in completed_frames:
                    found_files.append(expected_file)
                    continue

                # Try exact match first
                if os.path.exists(expected_file):
                    if os.path.getsize(expected_file) > 0:
                        found_files.append(expected_file)
                        if frame_id not in completed_frames:
                            newly_completed.append((frame_id, expected_file))
                            completed_frames.add(frame_id)
                else:
                    # Try pattern matching for alternative naming
                    base_name = os.path.basename(expected_file)
                    frame_part = base_name.split('_')[0] + '_' + base_name.split('_')[1]  # frame_XXXX_XXXXXX

                    patterns = [
                        os.path.join(self.output_folder, f"{frame_part}_*_clean_overlay.jpg"),
                        # os.path.join(self.output_folder, f"{frame_part}_*_occlusion_aware_overlay.jpg"),
                        # os.path.join(self.output_folder, f"{frame_part}_*_transparent.png")
                    ]

                    for pattern in patterns:
                        matches = glob.glob(pattern)
                        if matches:
                            match_file = matches[0]
                            if os.path.getsize(match_file) > 0:
                                found_files.append(match_file)
                                if frame_id not in completed_frames:
                                    newly_completed.append((frame_id, match_file))
                                    completed_frames.add(frame_id)
                                break

            # Report newly completed frames
            for frame_id, output_file in newly_completed:
                print(f"🎉 Frame {frame_id} completed! Output: {os.path.basename(output_file)}")

            # If we found outputs for all expected files, return them
            if len(found_files) == len(expected_files):
                print(f"✨ All {len(expected_files)} frames in batch completed!")
                return found_files

            # Check if process is still running
            if self.current_process and self.current_process.poll() is not None:
                print(f"🏁 Process finished with exit code: {self.current_process.poll()}")
                # Give it a moment for file system sync
                time.sleep(1)
                # Final check for remaining files
                final_files = self.final_file_check(expected_files)
                return final_files

            time.sleep(0.2)

        print(f"⏰ Timeout waiting for batch output after {timeout} seconds")
        print(f"📊 Completed frames: {len(completed_frames)}/{len(expected_files)}")
        return found_files if found_files else []

def main():
    # Configuration
    BATCH_SIZE = 12  # Process 12 frames per batch
    input_folder = '/content/CameraHMR-customized/input_frames'
    logo_path = '/content/CameraHMR-customized/demo_images/netflix_trans.png'
    output_folder = '/content/CameraHMR-customized/output_frames'

    # Initialize batch processor
    processor = BatchProcessor(input_folder, logo_path, output_folder, BATCH_SIZE)

    # Create directories
    os.makedirs(input_folder, exist_ok=True)
    os.makedirs(output_folder, exist_ok=True)

    # Clean up any existing files and processes
    processor.cleanup_folders()
    processor.kill_existing_processes()

    try:
        # Initialize webcam stream
        video_stream()

        frame_count = 0
        batch_frames = []  # Store frames for current batch
        batch_frame_paths = []  # Store frame file paths for current batch

        print("Starting batch-based webcam capture...")
        print(f"Batch size: {BATCH_SIZE} frames")

        while True:
            # Capture frame from webcam
            frames_in_batch = len(batch_frames)
            status = f"Batch {processor.batch_count + 1} - Collecting frame {frames_in_batch + 1}/{BATCH_SIZE}"

            js_reply = video_frame(status, '', '')

            if not js_reply:  # User clicked to stop
                break

            # Convert webcam image
            img = js_to_image(js_reply)

            # Add frame to current batch
            batch_frames.append(img)

            # Save input frame
            frame_filename = f'frame_{frame_count:04d}.jpg'
            input_path = os.path.join(input_folder, frame_filename)
            cv2.imwrite(input_path, img)
            batch_frame_paths.append(input_path)

            print(f"Collected frame {frame_count} for batch {processor.batch_count + 1}")

            frame_count += 1

            # Process batch when we have enough frames
            if len(batch_frames) >= BATCH_SIZE:
                print(f"\n=== Processing Batch {processor.batch_count + 1} ({len(batch_frames)} frames) ===")

                # Start processing script
                status = f"Batch {processor.batch_count + 1} - Starting processing script..."
                video_frame(status, '', '')

                process = processor.start_processing_script()

                # Wait for script to initialize
                print("Waiting for processing script to initialize...")
                time.sleep(5)

                # Check if process started successfully
                if process.poll() is not None:
                    print(f"ERROR: Processing script failed to start (exit code: {process.poll()})")
                    # Clear current batch and continue
                    batch_frames.clear()
                    batch_frame_paths.clear()
                    continue

                # Wait for all frames in the batch to be processed
                expected_outputs = []
                for i, frame_path in enumerate(batch_frame_paths):
                    base_name = os.path.basename(frame_path).replace('.jpg', '')
                    expected_output = os.path.join(output_folder, f'{base_name}_000000_clean_overlay.jpg')
                    expected_outputs.append(expected_output)

                status = f"Batch {processor.batch_count + 1} - Processing {len(batch_frames)} frames (may take 2-5 minutes)..."
                video_frame(status, '', '')

                # Wait for batch to complete (longer timeout for batches)
                timeout = 300 if processor.batch_count == 0 else 180  # 5 min for first batch, 3 min for others
                output_files = processor.wait_for_output(expected_outputs, timeout)

                if output_files:
                    print(f"Batch {processor.batch_count + 1} completed successfully!")

                    # Display results from the batch
                    for i, output_file in enumerate(output_files):
                        if os.path.exists(output_file):
                            try:
                                output_img = cv2.imread(output_file)
                                if output_img is not None:
                                    output_bytes = image_to_bytes(output_img)

                                    status = f"Batch {processor.batch_count + 1} - Displaying result {i + 1}/{len(output_files)}"
                                    video_frame(status, '', output_bytes)

                                    # Brief pause to show each result
                                    time.sleep(0.5)

                            except Exception as e:
                                print(f"Error displaying result {i}: {e}")

                    processor.total_frames_processed += len(batch_frames)

                    # Clean up batch files to save space
                    for frame_path in batch_frame_paths:
                        try:
                            if os.path.exists(frame_path):
                                os.remove(frame_path)
                        except:
                            pass

                    # Optional: Keep some outputs, remove others to save space
                    for output_file in output_files:
                        try:
                            # Keep only the final overlay, remove intermediate files
                            if 'clean_overlay' not in output_file:
                                os.remove(output_file)
                        except:
                            pass

                else:
                    print(f"Batch {processor.batch_count + 1} failed or timed out")

                # Terminate current process before starting next batch
                processor.terminate_current_process()

                # Clean up for next batch
                batch_frames.clear()
                batch_frame_paths.clear()
                processor.batch_count += 1

                # Brief pause between batches
                status = f"Batch {processor.batch_count} completed. Preparing for next batch..."
                video_frame(status, '', '')
                time.sleep(2)

                # Clean up output folder to prevent accumulation
                try:
                    for f in os.listdir(output_folder):
                        if f.endswith(('.png', '.jpg', '.obj', '.mtl')):
                            os.remove(os.path.join(output_folder, f))
                except:
                    pass

            # Small delay between frame captures
            time.sleep(0.1)

        # Process any remaining frames in incomplete batch
        if batch_frames:
            print(f"\nProcessing final incomplete batch of {len(batch_frames)} frames...")
            # Use same logic as above for final batch
            process = processor.start_processing_script()
            time.sleep(5)

            if process.poll() is None:
                expected_outputs = []
                for frame_path in batch_frame_paths:
                    base_name = os.path.basename(frame_path).replace('.jpg', '')
                    expected_output = os.path.join(output_folder, f'{base_name}_000000_clean_overlay.jpg')
                    expected_outputs.append(expected_output)

                output_files = processor.wait_for_output(expected_outputs, 180)

                if output_files:
                    print(f"Final batch completed successfully!")
                    processor.total_frames_processed += len(batch_frames)

    except KeyboardInterrupt:
        print("\nStopping due to keyboard interrupt...")
    except Exception as e:
        print(f"Error in main loop: {e}")
    finally:
        # Clean up
        print("\nCleaning up...")
        processor.terminate_current_process()
        processor.kill_existing_processes()

        print(f"Pipeline stopped. Total frames processed: {processor.total_frames_processed}")

if __name__ == "__main__":
    main()
```
